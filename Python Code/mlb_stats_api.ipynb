```python
#LEAGUE Param: 103 = American, 104 = National. 
#DIVISION Param: 200=AL West, 201=AL East, 202=AL Central, 203=NL West, 204=NL East, 205=NL Central


import pandas as pd

def get_division_standings(league, division):
    standings = pd.DataFrame(statsapi.standings_data(league)[division]['teams'])
    standings = standings[['name', 'div_rank', 'gb', 'w', 'l', 'elim_num', 'wc_rank', 'wc_gb', 'wc_elim_num']]
    return standings

mystandings = get_division_standings(103, 201)
```
