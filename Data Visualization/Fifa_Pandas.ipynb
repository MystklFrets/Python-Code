```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
style.use('fivethirtyeight')

#put the contents of the csv file into a variable object
soccer_stats = pd.read_csv('fifa19_kaggle.csv')

# select only first 5 rows
df = soccer_stats.head(10)

# select only listed columns
sd = df.reindex(columns=['Name', 'Age', 'Nationality', 'Overall', 'Club'])

# check dataset
print(sd)

sd.plot(kind='bar')

plt.title("fifa visual")
plt.xlabel("age")
plt.ylabel("skill")
plt.show()



```
